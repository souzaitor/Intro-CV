# Detect keypoints in an image using OpenCV

```python
%matplotlib inline
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('cameraman.tiff', 0)

plt.figure(figsize=[8,8])
plt.imshow(img, 'gray')

```

```python
sift = cv2.xfeatures2d.SIFT_create(nfeatures=200, nOctaveLayers=3, 
                                   contrastThreshold=0.04, 
                                   edgeThreshold=10, sigma=1.6)
kp = sift.detect(img, None)

img_keypoints = np.zeros_like(img)
outImg = cv2.drawKeypoints(img, kp, img_keypoints, 
                        flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.figure(figsize=[8,8])
plt.imshow(outImg)

```
